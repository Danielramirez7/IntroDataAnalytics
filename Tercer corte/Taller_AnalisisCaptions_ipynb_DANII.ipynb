## 🧹 Sección 2 (Limpieza)

Explica por qué es importante limpiar y normalizar el texto de los captions.  
Entregue un grafico de barras con las frecuencia en los captions

Limpiar y dejar parejo el texto de los captions es clave cuando se va a hacer análisis de texto, por varias razones:

Para que todo sea más consistente: A veces, cosas como las mayúsculas, la puntuación o las palabras mal escritas hacen que el sistema vea palabras iguales como si fueran diferentes. Al limpiar eso, se agrupan bien y se entiende mejor el contenido.

Para quitar lo que estorba: Se eliminan cosas que no aportan mucho al análisis, como los signos de puntuación, los números o palabras súper comunes como “el”, “la”, “y”, etc. Así uno se enfoca en lo que de verdad importa.

Para que los programas entiendan mejor: Muchos modelos de lenguaje o de inteligencia artificial necesitan que el texto esté bien organizado y limpio para poder funcionar bien.

Para tener mejores resultados: Entre más limpio esté el texto, más precisos son los análisis, como cuando uno quiere saber el sentimiento de una publicación, detectar temas, o entender de qué se está hablando en general.

```python
import matplotlib.pyplot as plt
from collections import Counter
import pandas as pd
import numpy as np

# Combina todo el texto limpio en una sola cadena
all_cleaned_text = ' '.join(df['cleaned_caption'].dropna())

# Divide el texto en palabras y cuenta la frecuencia
words = all_cleaned_text.split()
word_counts = Counter(words)
most_common_words = word_counts.most_common(20)

# DataFrame para graficar
most_common_df = pd.DataFrame(most_common_words, columns=['word', 'count'])

# Colores tipo Instagram
instagram_colors = ['#405DE6', '#5851DB', '#833AB4', '#C13584', '#E1306C',
                    '#FD1D1D', '#F56040', '#F77737', '#FCAF45', '#FFDC7D']

# Ajustar colores si hay más de 10 barras
colors = instagram_colors * (len(most_common_df) // len(instagram_colors) + 1)
colors = colors[:len(most_common_df)]

# Estilo personalizado del gráfico
plt.style.use('ggplot')
fig, ax = plt.subplots(figsize=(14, 7))
bars = ax.bar(most_common_df['word'], most_common_df['count'], color=colors, edgecolor='none')

# Etiquetas y título
ax.set_xlabel("Palabra", fontsize=14)
ax.set_ylabel("Frecuencia", fontsize=14)
ax.set_title("Top 20 palabras más usadas en captions (estilo Instagram)", fontsize=16, fontweight='bold')
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.yticks(fontsize=12)

# Mostrar valores sobre cada barra
for bar in bars:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2.0, yval + 1, int(yval),
            ha='center', va='bottom', fontsize=10, color='black')

# Quitar bordes del gráfico
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Fondo blanco
fig.patch.set_facecolor('white')
ax.set_facecolor('white')

# Añadir un pequeño crédito abajo
plt.figtext(0.99, 0.01, 'Fuente: Captions de Instagram', horizontalalignment='right', fontsize=10, color='gray')

plt.tight_layout()
plt.show()
```

```python
def classify_topics(text):
  prompt = (f"Clasifica por tematicas el siguiente caption:{text} "
              f"solo tienes entre estos temas juegos_desafios, promocion, personajes. solo responde con una palabra")
  respuesta = call_llm(prompt)
  return respuesta
```
